```python
import pandas as pd

# Load the regular season and playoff data (semicolon-separated and encoding handled)
acc_players_df = pd.read_csv('/Users/katetran/Documents/acc_players-2324F.csv', skiprows=1, encoding='latin-1')

#1 Display the first five rows of acc players dataset
print("ACC Players Stats:")
acc_players_df.head(5)
```

```python
# Rename columns for easier access
acc_players_df.rename(columns={'Pos': 'Position', 'MP' : 'Minutes Played', 'TRB' : 'Total Rebounds by Team', 
                              'AST' : 'Assists', 'STL' : 'Steals', 'BLK' : 'Blocked Shots', 'TOV' : 'Turnovers', 
                              'PTS' : 'Points', 'PER' : 'Player Efficiency Rating', 'WS' : 'Win Share'}, inplace=True)
acc_players_df.head(5)
```

```python
# Calculate the total points scored by all players combined.
total_points = acc_players_df['Points'].sum()

# Print total points
print("Total points scored by all players:")
print(total_points)
```

```python
# Find the player who has played the most minutes (MP).
most_minutes_played = acc_players_df[['Player', 'Minutes Played']].sort_values(by = 'Minutes Played', ascending=False).head(1)

# Print player with most minutes played
print("Player with most minutes:")
most_minutes_played.head()
```

```python
# Identify the top 5 players in terms of total rebounds (TRB).
top_total_rebounds = acc_players_df[['Player', 'Total Rebounds by Team']].sort_values(by = 'Total Rebounds by Team', ascending=False).head(5)

# Print top 5 players with most total rebounds
print('Top 5 Rebounders:')
top_total_rebounds.head()
```

```python
# Create a new DataFrame containing only players who played more than 500 minutes.
top_minutes_played_df = acc_players_df[acc_players_df['Minutes Played'] > 500]

print("New data frame containing only players who played more than 500 minutes created.")
top_minutes_played_df.head()
```

```python
# From this filtered DataFrame, determine the player with the highest total assists (AST).
highest_total_assists = top_minutes_played_df[['Player', 'Assists']].sort_values(by = 'Assists', ascending=False).head(1)

# Print player with highest total assists
print('Player with highest total assists')
highest_total_assists.head()
```

```python
# Who are the top 3 Assist Leads in the League?
top_assists = acc_players_df[['Player', 'Assists']].sort_values(by = 'Assists', ascending=False).head(3)

# Print top 3 assist leads in the league
print('Top 3 assist leads:')
top_assists.head()
```

```python
# Who are the top 3 Shot Blockers?
top_shot_blockers = acc_players_df[['Player', 'Blocked Shots']].sort_values(by = 'Blocked Shots', ascending=False).head(3)

# Print top 3 shot blockers in the league
print('Top 3 shot blockers:')
top_shot_blockers.head()
```

```python
# Group the players by School and calculate the total points scored by each school.
school_points = acc_players_df.groupby('School')['Points'].sum().reset_index()
school_points_sorted = school_points.sort_values(by = 'Points', ascending=False)
school_points_sorted.head()
```

```python
# Group the players by School and calculate the total assists (AST) for each team.
school_assists = acc_players_df.groupby('School')['Assists'].sum().reset_index()
school_assists_sorted = school_assists.sort_values(by = 'Assists', ascending=False)
school_assists_sorted.head()
```

```python
# Sort the schools by total points scored and display the top 3 schools.
school_points_sorted.head(3)
```

```python
# Create a bar chart showing the top 5 players by total points scored (PTS).
top_scorers = acc_players_df[['Player', 'Points']].sort_values(by = 'Points', ascending=False).head(5)

# Import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# Plot a bar chart of the top 5 players by total points
plt.figure(figsize=(10,6))
sns.barplot(x='Points', y='Player', data=top_scorers)
plt.title('Top 5 ACC Scorers')
plt.show()
```
